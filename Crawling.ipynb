```python
import requests
from bs4 import BeautifulSoup
from urllib.parse import urljoin
import pandas as pd
```

```python
def rating_transfer(rating):
    if rating > 5:
        rating = 1  # 1은 긍정적 리뷰
    elif rating <= 5:
        rating = 0  # 0은 부정적 리뷰
    pn = rating
    return pn
```

```python
review_list = []
rating_list = []
base_url = "https://www.imdb.com/"
key = ""

#수집하고 싶은 영화의 user riviews 페이지 url 붙여넣기
url = 'https://www.imdb.com/title/tt4154796/reviews/?ref_=tt_ql_2'
#수집하고 싶은 영화 리뷰 수 지정
MAX_CNT = 5000
cnt = 0

print("url = ", url)
res = requests.get(url) # 원격 API 호출
res.encoding = 'utf-8'
soup = BeautifulSoup(res.text, "lxml")

#평점 가져오기
for item in soup.select(".lister-list"):
    rating = item.select("span.rating-other-user-rating > span")
    if len(rating) == 2:
        rating = rating[0].text
    else:
        rating = ""
    review = item.select(".text")[0].text


load_more = soup.select(".load-more-data") #리뷰 더보기
flag = True

#리뷰가 남아있으면 url 생성
if len(load_more):
    ajaxurl = load_more[0]['data-ajaxurl']
    base_url = base_url + ajaxurl + "?ref_=undefined&paginationKey="
    key = load_more[0]['data-key']
else:
    flag = False  #리뷰가 더 없을 때

while flag:
    #생성한 url로 다시 리뷰 가져오기
    url = base_url + key
    print("url = ", url)
    res = requests.get(url)
    res.encoding = 'utf-8'
    soup = BeautifulSoup(res.text, "lxml")
    for item in soup.select(".lister-item-content"):
        rating = item.select("span.rating-other-user-rating > span")
        if len(rating) == 2:
            rating = rating[0].text
            review = item.select(".text")[0].text
            if rating_transfer(int(rating)) == 1:
                pn = "positive"   #평점이 5점 이상일 경우
            elif rating_transfer(int(rating)) == 0:
                pn = "negative"   #평점이 5점 미만일 경우
            review_list.append(review)
            rating_list.append(pn)
            cnt = cnt + 1
        else:
            rating = ""
        review = item.select(".text")[0].text   #리뷰 내용 추가

        if cnt >= MAX_CNT:
            break
    if cnt >= MAX_CNT:
        break
    load_more = soup.select(".load-more-data")
    if len(load_more):
        key = load_more[0]['data-key']
    else:
        flag = False
```

```python
df = pd.DataFrame(columns = ['review', 'sentiment'])
df['review'] = review_list
df['sentiment'] = rating_list
df.to_csv("IMDB_review.csv", index = False)
```
